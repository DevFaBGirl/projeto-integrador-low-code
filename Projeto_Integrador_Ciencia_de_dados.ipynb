```python
import pandas as pd


# Carregar dados
df = pd.read_csv("Dados/survey.csv")


# Traduzir colunas
colunas_traduzidas = {
    'Timestamp': 'Data',
    'Age': 'Idade',
    'Gender': 'Genero',
    'Country': 'Pais',
    'state': 'Estado',
    'self_employed': 'autônomo',
    'family_history': 'Historico_Familiar',
    'treatment': 'Tratamento',
    'work_interfere': 'Impacto_Trabalho',
    'no_employees': 'Quantidade_Funcionarios',
    'remote_work': 'Trabalho_Remoto',
    'tech_company': 'Empresa_Tecnologica',
    'benefits': 'Beneficios',
    'care_options': 'Opcoes_Cuidado',
    'wellness_program': 'Programa_Bem_Estar',
    'seek_help': 'Busca_Ajuda',
    'anonymity': 'Anonimato',
    'leave': 'Facilidade_Licenca',
    'mental_health_consequence': 'Consequencia_Saude_Mental',
    'phys_health_consequence': 'Consequencia_Saude_Fisica',
    'coworkers': 'Colegas',
    'supervisor': 'Supervisor',
    'mental_health_interview': 'Entrevista_Saude_Mental',
    'phys_health_interview': 'Entrevista_Saude_Fisica',
    'mental_vs_physical': 'Saude_Mental_vs_Fisica',
    'obs_consequence': 'Consequencia_Observada',
    'comments': 'Comentarios'
}


df = df.rename(columns=colunas_traduzidas)


# Traduzir respostas
df = df.replace({
    'Yes': 'Sim',
    'No': 'Não',
    'Maybe': 'Talvez',
    "Don't know": 'Não sei',
    'Not sure': 'Não tenho certeza',
    'Often': 'Muitas vezes',
    'Rarely': 'Raramente',
    'Sometimes': 'Às vezes',
    'Never': 'Nunca',
    'Somewhat easy': 'Um pouco fácil',
    'Very easy': 'Muito fácil',
    'Somewhat difficult': 'Um pouco difícil',
    'Very difficult': 'Muito difícil',
    'Some of them': 'Alguns deles'
})


# Manter apenas idades realistas
df = df[(df['Idade'] >= 18) & (df['Idade'] <= 120)]


# Preencher valores nulos por "Sem Informações"
df = df.fillna("Sem Informações")


# Corrigir quantidade de funcionários (evitar erro de data)
df['Quantidade_Funcionarios'] = df['Quantidade_Funcionarios'].replace({
    '1-5': '1 Até 5',
    '6-25': '6 até 25',
    '26-100': '26 até 100',
    '100-500': '100 até 500',
    '500-1000': '500 até 1000',
    'More than 1000': 'Mais de 1000'
})


# Traduzir gênero
df['Genero'] = df['Genero'].replace({

    # Masculino
    'Male': 'Masculino',
    'Cis Male': 'Masculino',
    'Cis Man': 'Masculino',
    'M': 'Masculino',
    'Male-ish': 'Masculino',
    'maile': 'Masculino',
    'Malr': 'Masculino',
    'Man': 'Masculino',
    'msle': 'Masculino',
    'Mail': 'Masculino',
    'cis male': 'Masculino',
    'Mal': 'Masculino',
    'Male (CIS)': 'Masculino',
    'Guy (-ish) ^_^': 'Masculino',
    'male': 'Masculino',
    'Male ': 'Masculino',
    'something kinda male?': 'Masculino',
    'ostensibly male, unsure what that really means': 'Masculino',
    'm': 'Masculino',
    'Make': 'Masculino',


    # Feminino
    'Female': 'Feminino',
    'Cis Female': 'Feminino',
    'cis-female/femme': 'Feminino',
    'F': 'Feminino',
    'femail': 'Feminino',
    'Woman': 'Feminino',
    'Femake': 'Feminino',
    'Female (cis)': 'Feminino',
    'female': 'Feminino',
    'woman': 'Feminino',
    'f': 'Feminino',
    'Female ': 'Feminino',


    # Outros / LGBTQIAPN+ / Não Binario
    'Agender': 'Outros',
    'Androgyne': 'Andrógina',
    'Female (trans)': 'Mulher-Trans',
    'fluid': 'Fluido',
    'Genderqueer': 'Queer',
    'queer': 'Queer',
    'non-binary': 'Não Binário',
    'queer/she/they': 'Queer',
    'Trans woman': 'Mulher-Trans',
    'Trans-female': 'Mulher-Trans',
    'Neuter': 'Outros',
    'Enby': 'Não Binário',
    'Nah': 'Não Binário',
    'All': 'Outros',
    'male leaning androgynous':'Andrógina'
})


# Traduzir países
df['Pais'] = df['Pais'].replace({
    'Australia': 'Austrália',
    'United States': 'Estados Unidos',
    'New Zealand': 'Nova Zelândia',
    'Austria': 'Áustria',
    'Canada': 'Canadá',
    'Denmark': 'Dinamarca',
    'Germany': 'Alemanha',
    'France': 'França',
    'India': 'Índia',
    'Italy': 'Itália',
    'Netherlands': 'Países Baixos',
    'Pakistan': 'Paquistão',
    'Philippines': 'Filipinas',
    'Poland': 'Polônia',
    'Singapore': 'Singapura',
    'South Africa': 'África do Sul',
    'Sweden': 'Suécia',
    'Switzerland': 'Suíça',
    'Belgium': 'Bélgica',
    'Bosnia and Herzegovina': 'Bósnia e Herzegovina',
    'Brazil': 'Brasil',
    'Bulgaria': 'Bulgária',
    'China': 'China',
    'Colombia': 'Colômbia',
    'Costa Rica': 'Costa Rica',
    'Croatia': 'Croácia',
    'Czech Republic': 'República Tcheca',
    'Finland': 'Finlândia',
    'Georgia': 'Geórgia',
    'Greece': 'Grécia',
    'Hungary': 'Hungria',
    'Ireland': 'Irlanda',
    'Israel': 'Israel',
    'Japan': 'Japão',
    'Latvia': 'Letônia',
    'Mexico': 'México',
    'Moldova': 'Moldávia',
    'Nigeria': 'Nigéria',
    'Norway': 'Noruega',
    'Portugal': 'Portugal',
    'Romania': 'Romênia',
    'Russia': 'Rússia',
    'Slovenia': 'Eslovênia',
    'Spain': 'Espanha',
    'Thailand': 'Tailândia',
    'United Kingdom': 'Reino Unido',
    'Uruguay': 'Uruguai'

})


# Corrigir Estado (apenas EUA mantém)
df.loc[df['Pais'] != 'Estados Unidos', 'Estado'] = 'Sem Informações'


# Converter a coluna Data para formato de data (datetime)
df['Data'] = pd.to_datetime(df['Data'])
df = df.sort_values(by='Data')
df['Data'] = df['Data'].dt.strftime('%d/%m/%Y')


# Remover coluna de comentários
df = df.drop(columns=['Comentarios'])


# Salvar pronto pro Excel
df.to_csv(
    "survey_Dados_Tratados.csv",
    index=False,
    sep=';',
    encoding='utf-8-sig'
)
```
